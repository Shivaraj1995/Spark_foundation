**Load the data**

```python
#Load the required libraries
import pandas as pd
import numpy as np
#To print the entire result 
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
#load the data
df_terrorism=pd.read_csv('globalterrorismdb_0718dist.csv',encoding = "ISO-8859-1")
#print first five records
df_terrorism.head(5)
```

```python
#print shape
print('Number of records:',df_terrorism.shape[0])
print('Number of Features:',df_terrorism.shape[1])
```

```python
#Print info
df_terrorism.info()
```

```python
#Check missing values
percent_miss=df_terrorism.isnull().sum()*100/len(df_terrorism)

miss_data=pd.DataFrame(percent_miss,columns=['Percent_of_missing'])
miss_data
```

```python
#Drop those columns which has missing value percent is more than 70%  (There is no any standard for dropping features so, here i consider 70% ).
dropped_feat=miss_data[~(miss_data['Percent_of_missing']>70)]
#print features names which has missing percent < 70%
dropped_feat.index
```

```python
#Load the data by considering only the above features

df_terrorism_v1=df_terrorism[dropped_feat.index]
df_terrorism_v1.head(5)
```

```python
#reset the index
df_terrorism_v1=df_terrorism_v1.reset_index(drop=False)
```

```python
df_terrorism_v1.isnull().sum()
```

```python
#Change the features names for better understanding
df_terrorism_v1.rename(columns={'iyear':'Year','imonth':'Month','iday':"day",'gname':'Group','country_txt':'Country','region_txt':'Region',
                     'provstate':'State','city':'City','latitude':'latitude','longitude':'longitude','summary':'summary',
                     'attacktype1_txt':'Attacktype','targtype1_txt':'Targettype','weaptype1_txt':'Weapon','nkill':'kill',
                     'nwound':'Wound'},inplace=True)
```

```python
#Consider only important features
df_terrorism_v2 = df_terrorism_v1[['Year','Month','day','Country','State','Region','City','latitude','longitude',"Attacktype",'kill',
               'Wound','target1','summary','Group','Targettype','Weapon']]
df_terrorism_v2.head(2)
```

**Exploratory Data Analysis**

General Analysis

```python
print("Country with most attacks: ",df_terrorism_v2["Country"].value_counts().idxmax())
print("City with most attacks: ",df_terrorism_v2["City"].value_counts().index[1])
print("Region with most attacks: ",df_terrorism_v2["Region"].value_counts().idxmax())
print("Year with most attacks: ",df_terrorism_v2["Year"].value_counts().idxmax())
print("Month with most attacks: ",df_terrorism_v2["Month"].value_counts().idxmax())
print("Group with most attacks: ",df_terrorism_v2["Group"].value_counts().index[1])
print("Most Attack Types: ",df_terrorism_v2["Attacktype"].value_counts().idxmax())
```

Find the year in which most attacks happened?

```python
#import library
import matplotlib.pyplot as plt
import seaborn as sns

year=df_terrorism_v2['Year'].unique()
years_count = df_terrorism_v2['Year'].value_counts(dropna = False).sort_index()
plt.figure(figsize=(15,10))
sns.barplot(x=year,y=years_count)
plt.xlabel('Year')
plt.ylabel('Number of attacks in each year')
plt.xticks(rotation=90)
plt.show()
```

The number of attacks more in 2014.

The Most affected Country?

```python
plt.figure(figsize = (15,7))
sns.barplot(x=df_terrorism_v2['Country'].value_counts()[:10].index,y=df_terrorism_v2['Country'].value_counts()[:10].values)
plt.title('Top Countries Affected')
plt.xlabel('Countries')
plt.ylabel('Count')
plt.xticks(rotation = 90)
plt.show()
```

The most affected country in terrorism is **Iraq**

Attack Groups

```python
plt.figure(figsize = (15,7))
sns.barplot(x =df_terrorism_v2['Group'].value_counts()[1:11].values, y = df_terrorism_v2['Group'].value_counts()[1:11].index,palette='magma')
plt.title('Top 10 Terrorist Organisation with Highest Terror Attacks')
plt.ylabel('Terror Group Name')
plt.xlabel('Number of Attacks')
plt.xticks(rotation= 90)
plt.show()
```

The Taliban Group is the top terrorist attack group

The most killed country?

```python
plt.figure(figsize = (15,7))
df_terrorism_v2.groupby(['Country'])['kill'].sum().sort_values(ascending = False).head(10).plot(kind = 'bar', colormap = 'coolwarm_r')
plt.xticks(rotation=90)
plt.title('No. of people killed Country wise')
plt.ylabel("Killed")
plt.show()
```

The most affected country from the Terrorism is Iraq

The less killed Countries

```python
#Less killed Countries 
plt.figure(figsize = (15,7))
df_terrorism_v2.groupby(['Country'])['kill'].sum().sort_values(ascending = False).tail(20).plot(kind = 'bar', colormap = 'Greens_r')
plt.xticks(rotation=90)
plt.title('No. of people killed Country wise')
plt.ylabel("Killed")
plt.show()
```

These countries are free from terrorism activities.

The number of injuries per year?

```python

plt.figure(figsize = (15,7))
df_terrorism_v2.groupby(['Year'])['Wound'].sum().sort_values(ascending = False).head(20).plot(kind = 'bar', colormap = 'Reds_r')
plt.xticks(rotation=90)
plt.title('No. of people Injured/yr')
plt.ylabel("Injured")
plt.show()
```

In 2015 most injured are happened due to terrorism activities.

Terrorist Activity

```python
pd.crosstab(df_terrorism_v2.Year, df_terrorism_v2.Region).plot(kind='area',stacked=False,figsize=(20,10))
plt.title('Terrorist Activities By Region In Each Year',fontsize=25)
plt.ylabel('Number of Attacks',fontsize=20)
plt.xlabel("Year",fontsize=20)
plt.show()
```

Attack Types

```python
plt.figure(figsize = (18,10))
df_terrorism_v2['Attacktype'].value_counts().plot.pie(autopct="%1.1f%%")
plt.title('Attack Types')
plt.show()
```

The Bombing/Explosion attacking type is more.

Target Types

```python
plt.figure(figsize = (18,12))
df_terrorism_v2['Targettype'].value_counts().plot.pie(autopct="%1.1f%%")
plt.title("Target_Types")
plt.show()
```

Targetted Person : Private Citzens & Property

Weapon Types

```python
plt.figure(figsize = (18,12))
df_terrorism_v2['Weapon'].value_counts().plot.pie(autopct="%1.1f%%")
plt.title("Weapon_Types")
plt.show()
```

The most commonly used weapons in terrorism activities are Explosion
